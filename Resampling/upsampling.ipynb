## Upsampling

Where you increase the frequency of the samples, such as from minutes to seconds.

```python
from pandas import read_csv
from datetime import datetime as dt
```

```python
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (20, 10)
```

```python
def parser(year_number):
    return dt.strptime(f'190{year_number}', '%Y-%m')
```

```python
series = read_csv('shampoo-sales.csv', 
                  header=0, index_col=0, parse_dates=True, squeeze=True, date_parser=parser)
upsampled = series.resample('D').mean()

print(series.head())
print(upsampled.head(32))
```

```python
series.plot()
```

```python
interpolated = upsampled.interpolate(method='linear')
print(interpolated.head(32))
interpolated.plot()
```

```python
# polinomial interpolation
interpolated = upsampled.interpolate(method='spline', order=2)
interpolated.plot()
```
