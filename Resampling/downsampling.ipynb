```python
from pandas import read_csv
from datetime import datetime as dt

from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (20, 10)
```

```python
def parser(year_number):
    return dt.strptime(f'190{year_number}', '%Y-%m')
```

```python
# Per quarter
series = read_csv('shampoo-sales.csv', 
                  header=0, index_col=0, parse_dates=True, squeeze=True, date_parser=parser)
resample = series.resample('Q').mean()
resample.plot()
```

```python
resample = series.resample('Y').sum()
resample.plot()
```
